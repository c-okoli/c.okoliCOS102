```python
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk

# Create main window
root = tk.Tk()
root.title("Delivery Service")
root.geometry("500x200")

# Create location label and entry
location_label = tk.Label(root, text="Location:")
location_label.pack()
location_entry = tk.Entry(root)
location_entry.pack()

# Create weight label and entry
weight_label = tk.Label(root, text="Weight in kg:")
weight_label.pack()
weight_entry = tk.Entry(root)
weight_entry.pack()

# Function to return cost based on location and weight
def calculate_cost(location,weight):
    if location == "Ibeju-Lekki":
        if weight < 10.0:
            return "N3,500"
        else:
            return "N5,000"
    elif location == "epe":
        if weight < 10.0:
            return "N5,000"
        else:
            return "N10,000"
    else:
        return "We do not deliver to your location."

# submit button callback function
def submit():
    location = location_entry.get().lower()
    weight_str = weight_entry.get()

# A try-except block to catch if weight input is wrong datatype
    try:
        weight = float(weight_str)
        cost = calculate_cost(location,weight)
        messagebox.showinfo("Payment Service", f"Total cost is {cost}")
    except:
        messagebox.showerror("Error", "Invalid weight, Please enter a valid number.")

# Create submit button
submit_button = tk.Button(root, text="Submit", command=submit)
submit_button.pack()

# Run the main event loop
root.mainloop()



        
        
        
        
```
